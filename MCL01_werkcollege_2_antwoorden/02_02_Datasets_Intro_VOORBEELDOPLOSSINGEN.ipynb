## VOORBEELD OPLOSSING A voor een wijn dataset


```python
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier

br = '\n'
data = load_wine()
keys = data.keys()
print (keys, br)

# get features and targets
X, y = data.data, data.target

# print dimensions and the first feature vector
print (f'features: {X.shape}')
print (f'targets : {y.shape}{br}')
print (f'X[0]:{br}{X[0]}{br}')

# print the names of features and target
features = data.feature_names
targets = data.target_names
print (f'feature set:{br}{features}{br}')
print (f'targets:{br}{targets}{br}')

# specifiy model and train model
rnd_clf = RandomForestClassifier(random_state=0, n_estimators=100)
rnd_clf.fit(X, y)

# print 6 most important features
rnd_name = rnd_clf.__class__.__name__  # name of model
feature_importances = rnd_clf.feature_importances_
importance = sorted(zip(feature_importances, features), reverse=True)
n = 6
print (n, 'most important features' + ' (' + rnd_name + '):')
[print (row) for i, row in enumerate(importance) if i < n]
```

```python
# load dataset
path = "../MCL01_werkcollege_2/data/"
datafile = path + "heights.csv"
heights = pd.read_csv(datafile)
# alternatief: heights = load_csv_pd_data(".././data/", "heights.csv")
heights
```

```python
# make a histogram of dataframe `heights` 
# using pandas `hist` function

# each column must be plotted,
# order is important for back and front graph
# both graphs are combined in one figure.
heights['Father'].hist(legend=True)
heights['Son'].hist(legend=True, alpha=0.3)
```

```python
# pandas.DataFrame.plot.hist() function
# which will plot the histogram of each column
# of the dataframe in the same figure. 
# Visibility is limited.
heights.plot.hist() 
heights.plot.hist(alpha=0.3)  # second plot 
```

```python
# Seaborn displot()
g = sns.displot(data=heights.melt(value_name="Height"),
                x="Height",
                col="variable"
               )
g.set_titles(col_template='{col_name}')
```

```python
# Seaborn histplot()

sns.displot(data=heights,
            bins=20,
            multiple='dodge',
           )
```

```python
# Seaborn scatterplot(): 
# plot heighs of Father against Son

sns.scatterplot(data=heights, 
                x="Father", y="Son")
```
